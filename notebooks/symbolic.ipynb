```python
from IPython.display import Math
import sympy as sym
from sympy.plotting import plot
sym.init_printing(order='none')
```

```python
x,y,h,c,k,A,B,C,D = sym.symbols("x y h c k A B C D")
```

```python
# Even Asphere
z = c*h**2/(1 + sym.sqrt(1-(1+k)*c**2*h**2)) + A*h**4 + B*h**6 + C*h**8

print("Even Polynomial Asphere")
print("sag:")
display(sym.Eq(sym.symbols('z'), z))

print("1st derivative:")
display( sym.Eq(sym.Derivative(sym.symbols('z'),h), sym.diff(z,h)) )

print("2nd derivative:")
display( sym.Eq(sym.Derivative(sym.symbols('z'),h,2), sym.diff(z,h,2)) )
```

```python
# Odd Asphere
z = c*h**2/(1 + sym.sqrt(1-(1+k)*c**2*h**2)) + A*h**3 + B*h**4 + C*h**5 + D*h**6

print("Odd Polynomial Asphere")
print("sag:")
display(sym.Eq(sym.symbols('z'), z))

print("1st derivative:")
display( sym.Eq(sym.Derivative(sym.symbols('z'),h), sym.diff(z,h)) )

print("2nd derivative:")
display( sym.Eq(sym.Derivative(sym.symbols('z'),h,2), sym.diff(z,h,2)) )
```
